```python
import astropy.units as u
import astropy.constants as const
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculateOrbitalVelocity(M, R):
  """
  Calculates the orbital velocity of a body around a central object.
  Input:
  M: Mass of the central object
  R: Distance from the central object
  Output:
  Orbital velocity of the body
  """
  return np.sqrt(const.G * M / R)
```

```python
# Find orbital velocity of an earth satellite
v = calculateOrbitalVelocity(1 * u.earthMass, 1 * u.earthRad)
print(v.to(u.km/u.s))
```

```python
# orbital distances for solar system

orbital_distances = np.array([
    0.387,  # Mercury
    0.723,  # Venus
    1.000,  # Earth
    1.524,  # Mars
    5.203,  # Jupiter
    9.537,  # Saturn
    19.191, # Uranus
    30.069  # Neptune
]) * u.AU
```

```python
# Use a vectorized application of the calculateOrbitalVelocity function to calculate the orbital velocities of the planets

orbital_velocities = calculateOrbitalVelocity(1 * u.Msun, orbital_distances)
print(orbital_velocities.to(u.km / u.s))
```

```python
# prompt: make a simple plot of orbital velocity against orbital distance

plt.plot(orbital_distances, orbital_velocities)
plt.xlabel('Orbital Distance (AU)')
plt.ylabel('Orbital Velocity (km/s)')
plt.title('Orbital Velocity vs. Distance from the Sun')
plt.show()
```

```python
# find rotation curve for milky way

orbital_distances = np.arange(1, 30) * 1000 * u.parsec # create list of orbital distances from 1 to 30 kiloparsecs
mass_blackhole = 4.3 * 10**6 * u.Msun # mass of central black hole
orbital_velocities = calculateOrbitalVelocity(mass_blackhole, orbital_distances) # calculate orbital velocities
plt.title('Milky Way Rotation Curve')
plt.xlabel('Orbital Distance (pc)')
plt.ylabel('Orbital Velocity (km/s)')
plt.plot(orbital_distances, orbital_velocities.to(u.km/u.s)) # plot rotation curve
```
